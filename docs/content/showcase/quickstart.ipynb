# Quickstart — FlowCite

Minimal example of registering items and tracking usage.


```python
from flowcite import register_item, bind, scoped_usage, track_item, report

register_item(
    id="example:2025:paper",
    type="article",
    title="An example paper",
    year=2025,
)
bind("mylib.analysis.run", ["example:2025:paper"])

@scoped_usage(target="mylib.analysis.run")
def run():
    track_item("example:2025:paper", used_by="mylib.analysis.run")

run()
print(report(format="markdown"))
```

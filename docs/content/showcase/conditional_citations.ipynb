# Conditional citations

This notebook shows how a function can decide which items to cite depending on its arguments.


```python
from flowcite import register_item, bind, scoped_usage, track_item, report

register_item(id="algo:alpha", type="article", title="Alpha algorithm paper")
register_item(id="algo:beta", type="article", title="Beta algorithm paper")

bind("mylib.run_method", ["algo:alpha", "algo:beta"])

@scoped_usage(target="mylib.run_method")
def run_method(mode="alpha"):
    if mode == "alpha":
        track_item("algo:alpha", used_by="mylib.run_method")
    else:
        track_item("algo:beta", used_by="mylib.run_method")

run_method("alpha")
print(report(format="markdown"))
```
